## üé® Task 7 ‚Äì Data Visualization & Insights

In this step, I‚Äôll create visualizations to better understand patterns and communicate insights from the Spotify dataset.

### Goals:
1. Identify which song features most influence popularity.
2. Visualize relationships between audio attributes such as energy, danceability, and valence.
3. Compare how song popularity varies across genres and other categories.

The visualizations below aim to make complex patterns in the data easier to interpret.


```python
# Popularity distribution
plt.figure(figsize=(8,5))
sns.histplot(df['popularity'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()

# Relationship between Energy and Popularity
plt.figure(figsize=(8,5))
sns.scatterplot(x='energy', y='popularity', data=df, alpha=0.4)
plt.title('Energy vs Popularity')
plt.xlabel('Energy')
plt.ylabel('Popularity')
plt.show()

# Relationship between Danceability and Popularity
plt.figure(figsize=(8,5))
sns.scatterplot(x='danceability', y='popularity', data=df, alpha=0.4, color='orange')
plt.title('Danceability vs Popularity')
plt.xlabel('Danceability')
plt.ylabel('Popularity')
plt.show()

# Top 10 genres by average danceability
genre_dance = df.groupby('track_genre')['danceability'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
genre_dance.plot(kind='bar', color='mediumseagreen')
plt.title('Top 10 Genres by Average Danceability')
plt.ylabel('Danceability')
plt.show()
```

### üß© Insights from Visualizations

- **Popularity Distribution:**  
  Most songs cluster around mid-range popularity scores, indicating that only a small portion of songs achieve very high popularity.

- **Energy vs. Popularity:**  
  There appears to be a weak positive relationship ‚Äî energetic songs tend to be slightly more popular.

- **Danceability vs. Popularity:**  
  Songs with moderate to high danceability often achieve higher popularity, suggesting upbeat tracks are favored.

- **Genre Insights:**  
  Certain genres, like pop and dance, tend to have higher average danceability, aligning with their strong presence in popular charts.

These visual insights help explain why certain song attributes correlate with Spotify popularity.


## üß† Task 6 ‚Äì Determine the Type of Analysis

The goal of this project is to explore how different audio features (such as danceability, energy, and tempo) affect a song‚Äôs popularity on Spotify.

Since this dataset contains numerical values (e.g., energy, loudness, valence) and categorical values (e.g., genre, explicit), the analysis will mainly focus on:

1. **Descriptive Analysis:**  
   To summarize key statistics (mean, median, etc.) and understand overall trends in the data.

2. **Correlation Analysis:**  
   To examine relationships between features (e.g., whether danceable songs tend to be more popular).  
   A heatmap will help visualize which variables are most strongly related to popularity.

3. **Comparative Analysis:**  
   To compare popularity across groups, such as:
   - **Genres** (e.g., which genre has the most popular songs)
   - **Explicit vs. Non-explicit songs**

4. **Visualization-Based Exploration:**  
   To communicate findings through intuitive charts like bar plots, box plots, and scatter plots.

This combination of analyses will help identify which song characteristics tend to make tracks more popular on Spotify.


```python
# Correlation Matrix
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=False)
plt.title('Correlation Matrix of Audio Features')
plt.show()

# Average popularity by genre
genre_popularity = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,5))
genre_popularity.head(10).plot(kind='bar', color='mediumseagreen')
plt.title('Top 10 Genres by Average Popularity')
plt.ylabel('Average Popularity')
plt.xlabel('Genre')
plt.show()

# Popularity vs Explicit
plt.figure(figsize=(6,4))
sns.boxplot(x='explicit', y='popularity', data=df)
plt.title('Popularity by Explicit Content')
plt.show()
```

```python
# Correlation matrix
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title('Correlation Matrix')
plt.show()

# Popularity by genre
genre_popularity = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False)
genre_popularity.head(10).plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title('Average Popularity by Genre')
plt.ylabel('Popularity')
plt.xlabel('Genre')
plt.show()

# Compare explicit vs non-explicit songs
sns.boxplot(x='explicit', y='popularity', data=df)
plt.title('Popularity vs Explicit Content')
plt.show()
```
